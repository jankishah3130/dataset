```python
x = df[['date','state','n_killed','n_injured']]
x['year'] = pd.to_datetime(df['date']).dt.year
x['month'] = pd.to_datetime(df['date']).dt.month
x['day'] = pd.to_datetime(df['date']).dt.day
x['day_of_week'] = pd.to_datetime(df['date']).dt.dayofweek
x['total_victim'] = x['n_killed'] + x['n_injured']
x.head()
```

```python
# on a particular date, how many person killed?
df[df['date'] == '2013-01-01']['n_killed'].sum()
```

```python
# on a particular date, how many person injured?
df[df['date'] == '2013-01-01']['n_injured'].sum()
```

```python
# for each date, how many person killed & injured?
df.groupby('date')[['n_killed','n_injured']].sum()
```

```python
df.groupby(x['year'])[['n_killed','n_injured']].sum()
# grouping the data according to the categories and applying a function to the categories.
```
